```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import ticker, cm
from matplotlib.ticker import FormatStrFormatter

# plt.rcParams['font.size'] = 14
# plt.rcParams['font.family'] = 'serif'
# fig = plt.figure(figsize=(6, 5))

r, z, E = np.loadtxt('micro.txt', unpack=True)

E_min = np.amin(E)
E_max = np.amax(E)

unq_r = np.unique(r)
unq_z = np.unique(z)
x, y = np.meshgrid(unq_r * 1e6, unq_z * 1e6)
z = np.reshape(E, (len(unq_r), len(unq_z)))

plt.contourf(x, y, z, cmap=plt.cm.jet, levels=np.linspace(E_min, E_max, 500))

CB = plt.colorbar(format='%.3f')
CB.set_label('Field Strength [kVcm$^{-1}$]')

plt.xlabel('y [$\mu$m]')
plt.ylabel('x [$\mu$m]')
plt.title('E-Field Near $r = 1$ $\mu$m Tips')

plt.tight_layout()
plt.show()

# plt.savefig('e_field_1_um_tips.pdf', bbox_inches='tight')
```

```python
plt.rcParams['font.size'] = 20
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
fig = plt.figure(figsize=(10, 8))

r, z, E = np.loadtxt('for_poster_E.txt', unpack=True)

E *= 1e-3

E_min = np.amin(E)
E_max = np.amax(E)

unq_r = np.unique(r)
unq_z = np.unique(z)
x, y = np.meshgrid(unq_r * 1e6, unq_z * 1e6)
z = np.reshape(E, (len(unq_z), len(unq_r)))

plt.contourf(x, y, z, cmap=plt.cm.jet, levels=np.linspace(E_min, E_max, 100), antialiased=True)
plt.contourf(-x, y, z, cmap=plt.cm.jet, levels=np.linspace(E_min, E_max, 100), antialiased=True)

CB = plt.colorbar(format='%i')
CB.set_label('Field Strength [MV/cm]')
CB.set_ticks([0, 1, 2, 3, 4, 5, 6])

r, z, V = np.loadtxt('for_poster_V.txt', unpack=True)

V_min = np.amin(V)
V_max = np.amax(V)

unq_r = np.unique(r)
unq_z = np.unique(z)
x, y = np.meshgrid(unq_r * 1e6, unq_z * 1e6)
z = np.reshape(V, (len(unq_z), len(unq_r)))

plt.contour(x, y, z, colors='#FFFFFF', alpha=0.5, levels=np.linspace(V_min, V_max, 15))
plt.contour(-x, y, z, colors='#FFFFFF', alpha=0.5, levels=np.linspace(V_min, V_max, 15))

plt.xlabel('r [$\mu$m]')
plt.ylabel('z [$\mu$m]')
plt.title('Field Near $r = 1$ $\mu$m Tip')

plt.gca().set_yticks([3499, 3500, 3501])
# plt.title('Electric Field Near $1$ $\mu$m Radius Tip\nw/ Bulk Field of 1 kVcm$^{-1}$')

plt.tight_layout()
# plt.show()

plt.savefig('e_field_large_tip.pdf', bbox_inches='tight')
```
