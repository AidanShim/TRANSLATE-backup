```python
from utils import Graph
import scipy.interpolate as interpolate
import glob
import re
from matplotlib import pyplot as plt
import numpy as np

database = 'Biagi'

P_v = [2, 7, 15]
N_v = [1.15e19, 4.02e19, 8.61e19]

def alpha(P,E,A,B):
    return P * A * np.exp(-B*P/E)

def reduced_alpha(x, A,B):
    return A * np.exp(-B/x)

def M(alpha,dx):
    return np.exp(alpha * dx)

def psi_to_torr(p):
    return p * 51.7149

def log_interp1d(xx, yy, kind='linear'):
    logx = np.log10(xx)
    logy = np.log10(yy)
    lin_interp = interpolate.interp1d(logx, logy, kind=kind)
    log_interp = lambda zz: np.power(10.0, lin_interp(np.log10(zz)))
    return log_interp

# This is just some helpful code to generate logarithmically evenly spaced V/cm
spacing = np.logspace(np.log10(1e2), np.log10(1e6), 12)
rounded = [float('%.2g'%i) for i in spacing]
for i in rounded:
    print(i, end=" ")
```

The first Townsend coefficient can be approximated by $\alpha = PAe^{-BP/E}$, where $A=14$ cm$^{-1}$ Torr$^{-1}$, $B=180$ V cm$^{-1}$ Torr$^{-1}$, $E$ is the electric field, and $P$ is the pressure (in Torr).

The total number of ionizations predicted by our simulation can be found by raising $2$ to the number of 'first hand' ionizations. When plotting the multiplication per $10$ micrometers, we use $\frac{M}{10\mu{m}} = 2^{\frac{10n}{\Delta{x}}}$ where $\Delta{x}$ is measured in micrometers.

Since $M = e^{\alpha\Delta{x}}$ (where in this paragraph $\Delta{x}$ is measured in centimeters), we can estimate $\alpha$ with $\frac{\ln{M}}{\Delta{x}}$, or $\frac{n\ln(2)}{\Delta{x}}$ where $n$ is the number of 'first hand' ionizations.

```python
from sympy.solvers import solve
from sympy import Symbol

# P = 13789.5 # Pa (2 PSI)
# P = 48263.3 # Pa (7 PSI)
P = 1.03421e5 # Pa (15 PSI)

V = 1e-6 # m^3
a = 1.355 * 1e5 * 1e-6 # Pa * m^6 / mol^2
b = 0.03201 * 1e-3 # m^3 / mol
R = 8.314 # J / (K * mol)
T = 87 # K

n = Symbol('n')

print(6.02214086e23 * solve( (a*b)/(V**2) * n**3 - (a/V) * n**2 + (P*b + R*T) * n - P*V, n)[0])
```

```python
file_tree = [[{}, {}, {}], [{}, {}, {}]]

E_vals = [[[], [], []], [[], [], []]]
alpha_vals = [[[], [], []], [[], [], []]]
M_vals = [[[], [], []], [[], [], []]]

v_vals = [[[], [], []], [[], [], []]]
v_devs = [[[], [], []], [[], [], []]]

eV_vals = [[[], [], []], [[], [], []]]
eV_devs = [[[], [], []], [[], [], []]]

colors = ['r', 'b', 'm']
colors_th = ['r--', 'b--', 'm--']

k = 1.81e-17
N = [1.15e19, 4.06e19, 8.81e19]

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,6))

for j, d in enumerate([1, 3]):
    for i, p in enumerate(P_v):
        for file in glob.glob('townsend-graphs/' + database + '/%i-psi/%id/*.txt'%(p,d)):
            key = int(re.search(r'(\d*)(?:V)', file).group(1))
            if (key in file_tree[j][i]):
                file_tree[j][i][key].append(file)
            else:
                file_tree[j][i][key] = [file]

        for key, value in sorted(file_tree[j][i].items()):
            group = Graph(value)
            E_vals[j][i].append(key)
            alpha_vals[j][i].append(group.alpha())
            M_vals[j][i].append(group.M())
            v_vals[j][i].append(group.drift_mean())
            v_devs[j][i].append(group.drift_std_dev())
            eV_vals[j][i].append(group.eV_mean())
            eV_devs[j][i].append(group.eV_std_dev())

        if d == 1:
            ax1.plot([v / psi_to_torr(p) for v in E_vals[j][i]], [a / psi_to_torr(p) for a in alpha_vals[j][i]], colors[i], label='%i PSI'%p)
        elif d == 3:
            ax2.plot([v / psi_to_torr(p) for v in E_vals[j][i]], [a / psi_to_torr(p) for a in alpha_vals[j][i]], colors[i], label='%i PSI'%p)
            # ax2.plot([v / psi_to_torr(p) for v in E_vals[j][i]], [k * N[i] * E / psi_to_torr(p) for E in eV_vals[j][i]])
    
x_v = np.logspace(1,3,100)

alpha_red_v_1 = reduced_alpha(x_v, 14, 180)
ax1.plot(x_v, alpha_red_v_1, 'g--', label=r'A = 14, B = 180')
ax2.plot(x_v, alpha_red_v_1, 'g--', label=r'A = 14, B = 180')

ax1.set_title(database + ' 1D')
ax2.set_title(database + ' 3D')

ax1.set_xlabel('Reduced Electric Field [V/cm/Torr]')
ax1.set_ylabel('Reduced $\\alpha$ [ions/cm/Torr]')

ax2.set_xlabel('Reduced Electric Field [V/cm/Torr]')
ax2.set_ylabel('Reduced $\\alpha$ [ions/cm/Torr]')

ax1.set_xlim(1e1, 1e3)
ax2.set_xlim(1e1, 1e3)

ax1.set_xscale('log')
ax1.set_yscale('log')

ax2.set_xscale('log')
ax2.set_yscale('log')

ax1.legend()
ax2.legend()

ax1.grid()
ax2.grid()

plt.show()
```

The above shows our predictions for the reduced Townsend coefficient versus the values expected from the Korff approximation.

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,6))

E_v = np.logspace(3, 6, 100)
for j, d in enumerate([1, 3]):
    for i, p in enumerate(P_v):
        a_v = alpha(psi_to_torr(p), E_v, 14, 180)
        mult_v = M(a_v, 10 * 1e-4)
        
        if d == 1:
            ax1.plot(E_vals[j][i], M_vals[j][i], colors[i], label='%i PSI'%p)
            ax1.plot(E_v, mult_v, colors_th[i], label=r'$e^{\alpha(%i)\Delta{x}}$'%p)
        elif d == 3:
            ax2.plot(E_vals[j][i], M_vals[j][i], colors[i], label='%i PSI'%p)
            ax2.plot(E_v, mult_v, colors_th[i], label=r'$e^{\alpha(%i)\Delta{x}}$'%p)

ax1.set_xlabel('Electric Field [V/cm]')
ax1.set_ylabel('Multiplication factor over 10 $\mu$m drift')

ax2.set_xlabel('Electric Field [V/cm]')
ax2.set_ylabel('Multiplication factor over 10 $\mu$m drift')

ax1.set_xscale('log')
ax1.set_yscale('log')

ax2.set_xscale('log')
ax2.set_yscale('log')

ax1.set_ylim(1, 1e5)
ax2.set_ylim(1, 1e5)

ax1.set_title(database + ' 1D')
ax2.set_title(database + ' 3D')

ax1.legend()
ax2.legend()

ax1.grid()
ax2.grid()

plt.show()
```

The above shows our simulation's predictions for charge multiplication versus the theoretical curve $M=e^{\alpha\Delta{x}}$.

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,6))

x = [1e-19, 3e-17, 2e-15]
y = [9e4, 3e5, 2e7]

data = log_interp1d(x, y)

x_vs = np.logspace(-19, np.log10(2e-15), 100)

for j, d in enumerate([1, 3]):
    for i, p in enumerate(P_v):
        if d == 1:
            ax1.errorbar([E / N_v[i] for E in E_vals[j][i]], [v * 1e2 for v in v_vals[j][i]], fmt=colors[i], yerr=[v * 1e2 for v in v_devs[j][i]], label='%i PSI'%p)
        elif d == 3:
            ax2.errorbar([E / N_v[i] for E in E_vals[j][i]], [v * 1e2 for v in v_vals[j][i]], fmt=colors[i], yerr=[v * 1e2 for v in v_devs[j][i]], label='%i PSI'%p)


ax1.plot(x_vs, data(x_vs), 'g--', label='Desired Fit')
ax2.plot(x_vs, data(x_vs), 'g--', label='Desired Fit')
            
ax1.set_xlabel('Reduced Electric Field [E/N]')
ax1.set_ylabel('Drift Velocity [cm/s]')

ax2.set_xlabel('Reduced Electric Field [E/N]')
ax2.set_ylabel('Drift Velocity [cm/s]')

ax1.set_xscale('log')
ax1.set_yscale('log')

ax1.set_xlim(1e-19, 1e-14)
ax1.set_ylim(1e4, 1e8)

ax2.set_xscale('log')
ax2.set_yscale('log')

ax2.set_xlim(1e-19, 1e-14)
ax2.set_ylim(1e4, 1e8)

ax1.legend()
ax2.legend()

ax1.set_title(database + ' 1D')
ax2.set_title(database + ' 3D')

ax1.grid()
ax2.grid()

plt.show()
```

The above plots simulated drift velocity versus measured drift velocity.
