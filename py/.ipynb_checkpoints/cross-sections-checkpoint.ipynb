```python
import numpy as np
import scipy.interpolate as interpolate
import math
import re
import glob
from matplotlib import pyplot as plt

m_e = 9.1e-31
database = 'Biagi'

# eV to m/s
def E_to_v(eV):
    return np.sqrt((2 * eV * 1.60218e-19) / (m_e))

# Low eV energy-independent LAr cross-section that merges with GAr cross section at 2.85 eV
def liquid_sigma_e(eV, gas_sigma_p):
    if eV < 3.85:
        return 5.5e-16
    return gas_sigma_p(eV)

# Low eV energy-independent LAr cross-section that merges with GAr cross section at 3.95 eV
def liquid_sigma_p(eV, S, gas_sigma_p):
    if eV < 4.6:
        return S(eV) * 5.5e-16
    return gas_sigma_p(eV)

sigma_p_eVs, sigma_p = np.loadtxt('cross-sections/' + database + '/momentum_xsec.txt', unpack=True)     # m^2
sigma_p *= 1e4                                                                             # changing to cm^2
sigma_p_func = interpolate.interp1d(sigma_p_eVs, sigma_p, kind='slinear', fill_value=(0., 0.), bounds_error=False)

sigma_i_eVs, sigma_i = np.loadtxt('cross-sections/' + database + '/ionization_xsec.txt', unpack=True)
sigma_i *= 1e4
sigma_i_func = interpolate.interp1d(sigma_i_eVs, sigma_i, kind='slinear', fill_value=(0., 0.), bounds_error=False)

# S is the structure factor of LAr
S_eVs = np.array([0, 1, 2, 3, 4, 5, 6, 7])
S = np.array([0.048, 0.055, 0.125, 0.4, 1.115, 1.29, 1.16, 1.01])
S_func = interpolate.interp1d(S_eVs, S, kind='quadratic', fill_value=(0.048, 1), bounds_error=False)

final_sigma_e = np.vectorize(lambda eV: liquid_sigma_e(eV, sigma_p_func))
final_sigma_p = np.vectorize(lambda eV: liquid_sigma_p(eV, S_func, sigma_p_func))

eVs_plot = np.logspace(-3, np.log10(950), 1000)
```

```python
fig = plt.figure(figsize=(16,8))

file_list_11 = glob.glob('cross-sections/' + database + '/11/excite_xsec_*.txt')
file_list_13 = glob.glob('cross-sections/' + database + '/13/excite_xsec_*.txt')
file_list_14 = glob.glob('cross-sections/' + database + '/14/excite_xsec_*.txt')
file_list_15 = glob.glob('cross-sections/' + database + '/15/excite_xsec_*.txt')

sigma_ex_11 = np.zeros(1000)
sigma_ex_13 = np.zeros(1000)
sigma_ex_14 = np.zeros(1000)
sigma_ex_15 = np.zeros(1000)

sigma_ex_tot = np.zeros(1000)

avg_11 = 0
for file in file_list_11:
    x_temp, y_temp = np.loadtxt(file, unpack=True)
    interp_func = interpolate.interp1d(x_temp, y_temp, kind='slinear', fill_value=(0., 0.), bounds_error=False)
    avg_11 += float(re.search(r'\d*\.\d*', file)[0])
    sigma_ex_11 += interp_func(eVs_plot) * 1e4
    sigma_ex_tot += interp_func(eVs_plot) * 1e4
avg_11 /= len(file_list_11)
    
avg_13 = 0
for file in file_list_13:
    x_temp, y_temp = np.loadtxt(file, unpack=True)
    interp_func = interpolate.interp1d(x_temp, y_temp, kind='slinear', fill_value=(0., 0.), bounds_error=False)
    avg_13 += float(re.search(r'\d*\.\d*', file)[0])
    sigma_ex_13 += interp_func(eVs_plot) * 1e4
    sigma_ex_tot += interp_func(eVs_plot) * 1e4
avg_13 /= len(file_list_13)
    
avg_14 = 0
for file in file_list_14:
    x_temp, y_temp = np.loadtxt(file, unpack=True)
    interp_func = interpolate.interp1d(x_temp, y_temp, kind='slinear', fill_value=(0., 0.), bounds_error=False)
    avg_14 += float(re.search(r'\d*\.\d*', file)[0])
    sigma_ex_14 += interp_func(eVs_plot) * 1e4
    sigma_ex_tot += interp_func(eVs_plot) * 1e4
avg_14 /= len(file_list_14)
    
avg_15 = 0
for file in file_list_15:
    x_temp, y_temp = np.loadtxt(file, unpack=True)
    interp_func = interpolate.interp1d(x_temp, y_temp, kind='slinear', fill_value=(0., 0.), bounds_error=False)
    avg_15 += float(re.search(r'\d*\.\d*', file)[0])
    sigma_ex_15 += interp_func(eVs_plot) * 1e4
    sigma_ex_tot += interp_func(eVs_plot) * 1e4
avg_15 /= len(file_list_15)

print("Average Excitation Values per Grouping\n", avg_11, avg_13, avg_14, avg_15)
    
plt.plot(eVs_plot, sigma_ex_11, label=avg_11)
plt.plot(eVs_plot, sigma_ex_13, label=avg_13)
plt.plot(eVs_plot, sigma_ex_14, label=avg_14)
plt.plot(eVs_plot, sigma_ex_15, label=avg_15)

plt.plot(eVs_plot, sigma_ex_tot)

plt.xlim(1e-3, 950)

plt.xlabel('Electron Energy [eV]')
plt.ylabel('Cross Section [cm$^2$]')

plt.xscale('log')
plt.yscale('log')

plt.legend()

plt.grid()
plt.show()
```

Above is a collection of Ar excitation cross-sections. These are taken from the BSR database on LXcat. In the simulation, we take the sum of these (the orange curve) to use as our excitation check. If an ionization occurs, our electron's energy is reduced by the mean excitation energy of the above graph: 13.62 eV.

```python
fig = plt.figure(figsize=(16,8))

E = 100e3
n = 1.15e19

plt.plot(eVs_plot, sigma_p_func(eVs_plot), label=r'GAr $\sigma_p$')
plt.plot(eVs_plot, final_sigma_e(eVs_plot), label=r'LAr $\sigma_E$')
plt.plot(eVs_plot, final_sigma_p(eVs_plot), label=r'LAr $\sigma_p$')

plt.plot(eVs_plot, sigma_i_func(eVs_plot), label=r'$\sigma_I$')
plt.plot(eVs_plot, sigma_ex_11, label=r'$\sigma_{ex,1}$')
plt.plot(eVs_plot, sigma_ex_13, label=r'$\sigma_{ex,2}$')
plt.plot(eVs_plot, sigma_ex_14, label=r'$\sigma_{ex,3}$')
plt.plot(eVs_plot, sigma_ex_15, label=r'$\sigma_{ex,4}$')

# plt.plot(eVs_plot, np.full(len(eVs_plot), E/n * 1e-2), label=r'$\sigma_{rec}$')

plt.xlabel('Electron Energy [eV]')
plt.ylabel(r'Cross Section [cm$^2$]')

plt.xscale('log')
plt.yscale('log')

plt.legend()
plt.grid()
plt.show()
```

Above is a graph of all cross-sections in use. GAr cross-sections are used for validation. LAr cross-sections have the form described by Wojcik and Tachiya.

```python
fig = plt.figure(figsize=(16,8))

compos_x_sec = np.array([])
for i, e in enumerate(eVs_plot):
    compos_x_sec = np.append(compos_x_sec, max(final_sigma_e(e), final_sigma_p(e)) + sigma_i_func(e) + sigma_ex_tot[i])

plt.plot(eVs_plot, sigma_p_func(eVs_plot) * E_to_v(eVs_plot) * 1e2, label=r'GAr $\sigma_p$')
plt.plot(eVs_plot, (sigma_p_func(eVs_plot) + sigma_i_func(eVs_plot) + sigma_ex_tot) * E_to_v(eVs_plot) * 1e2, label=r'GAr $\sigma_{total}$')

plt.plot(eVs_plot, final_sigma_e(eVs_plot) * E_to_v(eVs_plot) * 1e2, label=r'LAr $\sigma_E$')
plt.plot(eVs_plot, final_sigma_p(eVs_plot) * E_to_v(eVs_plot) * 1e2, label=r'LAr $\sigma_p$')
plt.plot(eVs_plot, compos_x_sec * E_to_v(eVs_plot) * 1e2, label=r'LAr $\sigma_{total}$')

plt.plot(eVs_plot, sigma_i_func(eVs_plot) * E_to_v(eVs_plot) * 1e2, label=r'$\sigma_I$')
plt.plot(eVs_plot, sigma_ex_tot * E_to_v(eVs_plot) * 1e2, label=r'$\sigma_{ex}$')

print(np.amax(compos_x_sec * E_to_v(eVs_plot) * 1e2))

plt.xlabel('Electron Energy [eV]')
plt.ylabel(r'Velocity $\times$ Cross Section [cm$^3$/s]')

plt.xscale('log')
plt.yscale('log')

plt.legend()
plt.grid()
plt.show()
```

We check whether an event occurs by evaluating $\frac{v\sigma(v)}{K_{max}}$. Wojcik and Tachiya use a velocity-independent $K_{max}$, which we see must exceed $10^{-6}$. Alternatively, one can use a velocity-dependent $K_{max}$ (which we do for the LAr simulation, but not for GAr) proportional to the above $\sigma_{total}$ (as $K_{max}$ controls both collision probability and time step length).

```python
eVs_save = np.logspace(-3, np.log10(950), 1000)

roof = 5

sigma_e_log = np.array([])
sigma_p_log = np.array([])
sigma_i_log = np.array([])
K_max_log = np.array([])

np.savetxt('gen-cross-sections/' + database + '/sigma_p_gas.txt', sigma_p_func(eVs_plot), newline=',')
np.savetxt('gen-cross-sections/' + database + '/K_max_gas.txt', (sigma_p_func(eVs_plot) + sigma_i_func(eVs_plot) + sigma_ex_tot) * E_to_v(eVs_plot) * 1e2 * 1e2 * roof, newline=',')

np.savetxt('gen-cross-sections/' + database + '/sigma_e_liquid.txt', final_sigma_e(eVs_plot), newline=',')
np.savetxt('gen-cross-sections/' + database + '/sigma_p_liquid.txt', final_sigma_p(eVs_plot), newline=',')
np.savetxt('gen-cross-sections/' + database + '/K_max_liquid.txt', compos_x_sec * E_to_v(eVs_plot) * 1e2 * roof, newline=',')

np.savetxt('gen-cross-sections/' + database + '/sigma_i.txt', sigma_i_func(eVs_plot), newline=',')
np.savetxt('gen-cross-sections/' + database + '/sigma_excite_11.txt', sigma_ex_11, newline=',')
np.savetxt('gen-cross-sections/' + database + '/sigma_excite_13.txt', sigma_ex_13, newline=',')
np.savetxt('gen-cross-sections/' + database + '/sigma_excite_14.txt', sigma_ex_14, newline=',')
np.savetxt('gen-cross-sections/' + database + '/sigma_excite_15.txt', sigma_ex_15, newline=',')
```
