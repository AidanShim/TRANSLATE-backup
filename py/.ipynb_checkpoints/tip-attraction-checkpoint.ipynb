```python
import glob
import re
from matplotlib import pyplot as plt
from utils import Graph
import numpy as np
from matplotlib import cm
from collections import OrderedDict

def nano_tip_plot(graph_start, graph_end):
    x = np.arange(graph_start, graph_end + 1, 1)
    
    tip_x = []
    tip_y = []
    
    for i in x:
        tip_x.append(-0.15001 + i)
        tip_x.append(-0.15 + i)
        tip_x.append(0.15 + i)
        tip_x.append(0.15001 + i)
        tip_y.append(0)
        tip_y.append(0.6)
        tip_y.append(0.6)
        tip_y.append(0)
        
    return tip_x, tip_y

def micro_tip_plot(graph_start, graph_end):
    x = np.arange(graph_start, graph_end + 1, 1)
    
    tip_x = []
    tip_y = []
    
    for i in x:
        tip_x.append(-1.001 + i)
        tip_x.append(-1 + i)
        tip_x.append(1 + i)
        tip_x.append(1.001 + i)
        tip_y.append(0)
        tip_y.append(3.5)
        tip_y.append(3.5)
        tip_y.append(0)
        
    return tip_x, tip_y

def hit_check(x, y, z):
    h = x
    
    if h >= 0.0:
        return True
    
    return False

def mil_hit_check(x, y, z):
    r = (y**2 + z**2)**0.5
    h = x
    
    if (h <= 5e2) and (h <= (-5.05050505 * (r - 1) + 501)):
        return True
    
    return False
```

```python
attraction_ratio = []

file_tree = {}

for file in glob.glob('tip-attraction/micrometer/single/*.txt'):
    key = int(re.search(r'(\d*)(?:V)', file).group(1))
    if (key in file_tree):
        file_tree[key].append(file)
    else:
        file_tree[key] = [file]

for key, value in sorted(file_tree.items()):
    group = Graph(value, smooth=False, cutoff=0.01, order=3)

    fig = plt.figure(figsize=(16,8))

    hits = 0

    for i in range(group.n):
        x, z = group.x[i], group.z[i]
        if hit_check(x[-1], group.y[i][-1], z[-1]) is True:
            hits += 1
            plt.plot(z, x, color=cm.plasma(np.log(group.ionized[i][-1] / (max_ionized + 1) * np.exp(1) + 1)), alpha=(1 - abs(group.y[i][-1])/10), label=int(group.ionized[i][-1]))
        else:
            plt.plot(z, x, color='#BBBBBB', alpha=(1 - abs(group.y[i][-1])/10))

    attraction_ratio.append(hits / group.n)

    x_tips, y_tips = micro_tip_plot(0, 0)
    plt.plot(x_tips, y_tips, color='0.0')

    plt.xlabel('z Position [$\mu$m]')
    plt.ylabel('x Position [$\mu$m]')
    plt.title('Electrons spawned randomly in disc of $r = 10 \mu$m, $80 \mu$m above tip')

    plt.xlim(-80, 80)
    plt.ylim(0, 80)

    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip([int(j) for j in labels], handles))
    sorted_legend = dict(sorted(by_label.items(), reverse=True))
    plt.legend(sorted_legend.values(), sorted_legend.keys())

    plt.show()

    fig = plt.figure(figsize=(16,8))
    plt.hist(group.ke, bins='auto', rwidth=1.0)

    plt.xlabel('Energy [eV]')
    plt.ylabel('Instances')
    plt.title('Energy Levels Occupied During Simulations')

    plt.show()
```

```python
file_tree = {}

for file in glob.glob('tip-attraction/single/single/*.txt'):
    key = int(re.search(r'(\d*)(?:V)', file).group(1))
    if (key in file_tree):
        file_tree[key].append(file)
    else:
        file_tree[key] = [file]

for key, value in sorted(file_tree.items()):
    group = Graph(value, smooth=False, cutoff=0.01, order=3)

    ionization_list = [k[-1] for k in group.ionized]
    max_ionized = np.max([k[-1] for k in group.ionized])

    fig = plt.figure(figsize=(16,60))
    
    hits = 0

    for i in range(group.n):
        x, z = group.x[i], group.z[i]
        if mil_hit_check(x[-1], group.y[i][-1], z[-1]) is True:
            hits += 1
            plt.plot(z, x, color=cm.plasma(np.log(group.ionized[i][-1] / (max_ionized + 1) * np.exp(1) + 1)), alpha=(1 - abs(group.y[i][-1])/10), label=int(group.ionized[i][-1]))
        else:
            plt.plot(z, x, color='#BBBBBB', alpha=(1 - abs(group.y[i][-1])/10))

    attraction_ratio.append(hits / group.n)
        
    plt.plot([-100, -1, 1, 100], [0, 500, 500, 0], color='0.0')

    plt.xlabel('z Position [$\mu$m]')
    plt.ylabel('x Position [$\mu$m]')
    plt.title('Electrons spawned randomly in disc of $r = 100 \mu$m, $250 \mu$m above tip')

    plt.xlim(-100, 100)
    plt.ylim(0, 750)

    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip([int(j) for j in labels], handles))
    sorted_legend = dict(sorted(by_label.items(), reverse=True))
    plt.legend(sorted_legend.values(), sorted_legend.keys())
    
    plt.show()
    
    fig = plt.figure(figsize=(16,8))
    plt.hist(group.ke, bins='auto', rwidth=1.0)

    plt.xlabel('Energy [eV]')
    plt.ylabel('Instances')
    plt.title('Energy Levels Occupied During Simulations')

    plt.show()
    
    fig = plt.figure(figsize=(16,8))
    plt.hist(ionization_list, bins=100)
    
    plt.xlabel('Primary Ionizations')
    plt.ylabel('Simulations Ending with x Primary Ionizations')
    
    plt.show()
```
