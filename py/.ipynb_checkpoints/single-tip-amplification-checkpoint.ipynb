```python
import glob
import re
from matplotlib import pyplot as plt
from utils import Graph
import numpy as np
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from collections import OrderedDict

def hit_check(x, y, z):
    h = x
    
    if h >= 0.0:
        return True
    
    return False
```

```python
file_tree = {}

for file in glob.glob('simulation-runs/3d/*.txt'):
    key = int(re.search(r'(\d*)(?:V)', file).group(1))
    if (key in file_tree):
        file_tree[key].append(file)
    else:
        file_tree[key] = [file]

for key, value in sorted(file_tree.items()):
    group = Graph(value, smooth=False, cutoff=0.01, order=3)

    ionization_list = [k[-1] for k in group.ionized]
    max_ionized = np.max([k[-1] for k in group.ionized])
    multiplication_factor = sum([2**k for k in ionization_list]) / group.n

    fig = plt.figure(figsize=(16,32))
    
    hits = 0

    for i in range(group.n):
        x, y, z = group.x[i], group.y[i], group.z[i]
        print(group.x[i][np.where(group.ionized[i] == 1)[0]])
        if hit_check(x[-1], y[-1], z[-1]) is True:
            hits += 1
            plt.plot(y, x, color=cm.plasma(group.ionized[i][-1] / max_ionized), alpha=0.6, label=int(group.ionized[i][-1]))
        else:
            plt.plot(y, x, color='#BBBBBB', alpha=0.6)
        
    plt.plot([-1e-4 * 1e6, -1e-7 * 1e6, -0.5e-7 * 1e6, 0, 0.5e-7 * 1e6, 1e-7 * 1e6, 1e-4 * 1e6], [3e-3 * 1e6, 3.5e-3 * 1e6, (3.5e-3 + 0.866025404e-7) * 1e6, (3.5e-3 + 1e-7) * 1e6, (3.5e-3 + 0.866025404e-7) * 1e6, 3.5e-3 * 1e6, 3e-3 * 1e6], color='0.0')

    plt.xlabel('z Position [$\mu$m]')
    plt.ylabel('x Position [$\mu$m]')
    plt.title('Single Millimeter Tip - Electrons spawned randomly in disc of $r = 3.5 \mu$m, $40 \mu$m above tip (%i V/cm) (%0.2f attraction) (%0.2f mult factor)'%(key, hits / group.n, multiplication_factor))

    plt.xlim(-1e-5 * 1e6, 1e-5 * 1e6)
    plt.ylim(3.49e-3 * 1e6, 3.51e-3 * 1e6)

    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip([int(j) for j in labels], handles))
    sorted_legend = dict(sorted(by_label.items(), reverse=True))
    plt.legend(sorted_legend.values(), sorted_legend.keys())
    
    plt.show()
    
    fig = plt.figure(figsize=(16,8))
    
    plt.hist([k[-200:] for k in group.ke], bins=50, rwidth=1.0)
    
    plt.ylim(0, 80)

    plt.xlabel('Energy [eV]')
    plt.ylabel('Instances')
    plt.title('Single Millimeter Tip - Energy Levels Occupied During Simulations (%0.2f mean, %0.2f standard deviation)'%(group.eV_mean(200), group.eV_std_dev(200)))

    plt.show()
```
