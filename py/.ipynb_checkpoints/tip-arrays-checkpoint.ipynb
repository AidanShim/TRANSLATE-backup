```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt
from utils import Graph

size_shape = 'micrometer'

x, y, E_x, E_y = np.loadtxt('tip-arrays/' + size_shape + '/comsol_data.txt', unpack=True)

strength_map = np.empty([301, 101])
E_x_map = np.empty([301, 101])
E_y_map = np.empty([301, 101])

for j in range(301):
    for i in range(101):
        strength_map[j][i] = sqrt(E_x[i + 101 * j]**2 + E_y[i + 101 * j]**2)
        E_x_map[j][i] = E_x[i + 101 * j]
        E_y_map[j][i] = E_y[i + 101 * j]
        
plt.imshow(np.append(np.append(E_x_map, E_y_map, axis=1), strength_map, axis=1), origin='lower', interpolation='bicubic')

np.savetxt('gen-tip-arrays/' + size_shape + '/E_x.txt', E_x, newline=',')
np.savetxt('gen-tip-arrays/' + size_shape + '/E_y.txt', E_y, newline=',')
```

```python
import glob
import re

file_tree = [{}, {}]

for i, d in enumerate([1, 3]):
    for file in glob.glob('simulation-runs/%id/*.txt'%d):
        key = int(re.search(r'(\d*)(?:V)', file).group(1))
        if (key in file_tree[i]):
            file_tree[i][key].append(file)
        else:
            file_tree[i][key] = [file]

for key, value in sorted(file_tree[1].items()):
    fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(16,4))
    
    group = Graph(value)
    
    for i in range(group.n):
        x, y, z = group.x[i], group.y[i], group.z[i]
        ax1.plot(x, y)
        ax2.plot(x, z)
        ax3.plot(y, z)
        
    ax1.set_xlabel('x [$\mu$m]')
    ax1.set_ylabel('y [$\mu$m]')
    
    #ax1.set_xlim(0, 7)
    #ax1.set_ylim(-8, 8)

    ax2.set_xlabel('x [$\mu$m]')
    ax2.set_ylabel('z [$\mu$m]')
    
    #ax2.set_xlim(0, 7)
    #ax2.set_ylim(-8, 8)
    
    ax3.set_xlabel('y [$\mu$m]')
    ax3.set_ylabel('z [$\mu$m]')
    
    #ax3.set_xlim(-8, 8)
    #ax3.set_ylim(-8, 8)
    
    plt.show()
```

```python
for key, value in sorted(file_tree[1].items()):
    fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(16,4))
    
    group = Graph(value)
    
    for i in range(group.n):
        x, y, z, t = group.x[i], group.y[i], group.z[i], group.t[i]
        ax1.plot(t, x)
        ax2.plot(t, y)
        ax3.plot(t, z)
        
    ax1.set_xlabel('t [ns]')
    ax1.set_ylabel('x [$\mu$m]')

    ax2.set_xlabel('t [ns]')
    ax2.set_ylabel('y [$\mu$m]')
    
    ax3.set_xlabel('t [ns]')
    ax3.set_ylabel('z [$\mu$m]')
    
    plt.show()
```
