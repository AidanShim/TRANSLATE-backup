```python
import numpy as np
from matplotlib import pyplot as plt
import glob
import re
from utils import Graph

file_tree = [{}, {}]
colors = ['r', 'b']

for i, d in enumerate([1, 3]):
    for file in glob.glob('diffusion/%id/*.txt'%d):
        key = int(re.search(r'(\d*)(?:V)', file).group(1))
        if (key in file_tree[i]):
            file_tree[i][key].append(file)
        else:
            file_tree[i][key] = [file]

    for key, value in sorted(file_tree[i].items()):
        group = Graph(value)
        
        for k in range(14):
            x = []
            y = []
            z = []
            for j in range(group.n):
                x.append(group.x[j][k * 100])
                y.append(group.y[j][k * 100])
                z.append(group.z[j][k * 100])
            fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(16,4))
            
            ax1.hist(x, color=colors[i], rwidth=0.9)
            ax1.set_xlim(-6, 6)
            ax1.set_title('%iD - x-coordinate at '%d + str(group.t[0][k]) + ' ns')
            ax1.set_xlabel('x [$\mu$m]')
            ax1.set_ylabel('Number of Electrons')
            
            ax2.hist(y, color=colors[i], rwidth=0.9)
            ax2.set_xlim(-6, 6)
            ax2.set_title('%iD - y-coordinate at '%d + str(group.t[0][k]) + ' ns')
            ax2.set_xlabel('y [$\mu$m]')
            ax2.set_ylabel('Number of Electrons')
            
            ax3.hist(z, color=colors[i], rwidth=0.9)
            ax3.set_xlim(-6, 6)
            ax3.set_title('%iD - z-coordinate at '%d + str(group.t[0][k]) + ' ns')
            ax3.set_xlabel('z [$\mu$m]')
            ax3.set_ylabel('Number of Electrons')

            plt.show()
        
```
