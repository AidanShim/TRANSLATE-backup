```python
import glob
import cycler

import numpy as np
import tikzplotlib
from matplotlib import pyplot as plt
import matplotlib as mpl

# plt.rcParams['font.size'] = 14
# plt.rcParams['font.family'] = 'serif'
# fig = plt.figure(figsize=(6, 6))

est_vals = [[[], [], []], [[], [], []]]

for j, d in enumerate([1, 3]):
    x, y = [], []
    for i, m in enumerate([10, 100, 1000]):
        ionized = []
        child_ions = []
        est_ionized = []
        files = 0
        for file in glob.glob('ion-est-uni-fields/%i/%id/*.txt'%(m,d)):
            _t, _x, _y, _z, _ke, _drift, _ionized, _child_ions = np.loadtxt(file, delimiter=',', unpack=True, ndmin=1)
            est_vals[j][i].append(2**_child_ions)
            ionized.append(_ionized)
            files += 1
        mu = np.mean(np.log(est_vals[j][i]))
        var = np.var(np.log(est_vals[j][i]))
        avg_est_ionized = np.exp(mu + 0.5 * var)
        avg_ionized = np.mean(ionized)
        x.append(avg_ionized)
        y.append(avg_est_ionized)
    plt.plot(x, y, label=f'Log-Normal Mean ({d}D)', color=f'C{j}')

x = np.logspace(1, 3, 10)

plt.xscale('log')
plt.yscale('log')

plt.xlabel('Actual Number of Ionizations')
plt.ylabel('Estimated Number of Ionizations')
plt.title('Ionizations in a Uniform Field')

plt.plot(x, x, label='Desired Fit', color='C2')

plt.legend()
plt.tight_layout()
plt.grid()
plt.show()

# plt.savefig('../../../documentation/plots/ionizations_in_uni_field.pdf', bbox_inches='tight')
```
