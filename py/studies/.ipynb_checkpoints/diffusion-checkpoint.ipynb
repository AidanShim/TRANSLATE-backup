```python
import glob
import re
from math import sqrt, floor, ceil

import numpy as np
from scipy import stats

from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

from utils import Graph

colors = ['C0', 'C1']
field_strength = 2000
```

```python
file_tree = [{}, {}]

for i, d in enumerate([1, 3]):
    for file in glob.glob('diffusion/%id/%i/*.txt'%(d, field_strength)):
        key = int(re.search(r'(\d*)(?:V)', file).group(1))
        if (key in file_tree[i]):
            file_tree[i][key].append(file)
        else:
            file_tree[i][key] = [file]
```

```python
threshold = 3

for i, d in enumerate([1, 3]):
    for key, value in sorted(file_tree[i].items()):
        group = Graph(value)
        
        min_x_dist = np.amin(([np.amax(np.floor(k)) for k in group.x])) # The greatest x value all simulations have in common
        x_index = [np.where(np.floor(k)==min_x_dist)[0][0] for k in group.x] # The data lines where these x values first occur
        
        t_vals = [group.t[k][x_index[k]] for k in range(group.n)]
        t_vals = [t_vals[k] for k in np.where(np.abs(stats.zscore(t_vals)) < threshold)[0]]
        
        v_vals = [(group.x[k][x_index[k]] / group.t[k][x_index[k]] * 1e3) for k in range(group.n)]
        v_vals = [v_vals[k] for k in np.where(np.abs(stats.zscore(t_vals)) < threshold)[0]]
        
        t_var = np.var(t_vals, ddof=1)
        v_avg = np.mean(v_vals)
        
        print('Our V_d:', int(v_avg))
        print('Their V_d:', 1648, '\n')
        
        print('D_L measured with time spread (our V_d):', (t_var / min_x_dist) * v_avg**3 * 0.5 * 1e-8)
        print('D_L measured with time spread (their V_d):', (t_var / min_x_dist) * 1648**3 * 0.5 * 1e-8, '\n')
        
        x_vals = [group.x[k][-1] for k in range(group.n)]
        y_vals = [group.y[k][-1] for k in range(group.n)]
        z_vals = [group.z[k][-1] for k in range(group.n)]
        
        print('D_L measured with position spread:', np.var([x_vals[k] for k in np.where(np.abs(stats.zscore(x_vals)) < threshold)[0]], ddof=1) / 20 * 1e1)
        print('D_Ty measured with position spread:', np.var([y_vals[k] for k in np.where(np.abs(stats.zscore(y_vals)) < threshold)[0]], ddof=1) / 20 * 1e1)
        print('D_Tz measured with position spread:', np.var([z_vals[k] for k in np.where(np.abs(stats.zscore(z_vals)) < threshold)[0]], ddof=1) / 20 * 1e1)
```

```python
threshold = 3

for i, d in enumerate([1, 3]):
    for key, value in sorted(file_tree[i].items()):
        group = Graph(value)
        
        fig = plt.figure(figsize=(16,8))
        
        sigma_squared_L = []
        sigma_squared_T1 = []
        sigma_squared_T2 = []
        t = []
        
        for k in range(0, len(group.t[0]), round(len(group.t[0]) / 100)):
            t_vals = [group.t[l][k] for l in range(group.n)]
            x_vals = [group.x[l][k] for l in range(group.n)]
            y_vals = [group.y[l][k] for l in range(group.n)]
            z_vals = [group.z[l][k] for l in range(group.n)]
            t.append(np.mean(t_vals) * 1e-9)
            sigma_squared_L.append(np.var([x_vals[k] for k in np.where(np.abs(stats.zscore(x_vals)) < threshold)[0]], ddof=1) * 1e-8)
            sigma_squared_T1.append(np.var([y_vals[k] for k in np.where(np.abs(stats.zscore(y_vals)) < threshold)[0]], ddof=1) * 1e-8)
            sigma_squared_T2.append(np.var([z_vals[k] for k in np.where(np.abs(stats.zscore(z_vals)) < threshold)[0]], ddof=1) * 1e-8)
        
        slope_L, intercept_L, r_value_L, p_value_L, std_err_L = stats.linregress(t, sigma_squared_L)
        slope_T1, intercept_T1, r_value_T1, p_value_T1, std_err_T1 = stats.linregress(t, sigma_squared_T1)
        slope_T2, intercept_T2, r_value_T2, p_value_T2, std_err_T2 = stats.linregress(t, sigma_squared_T2)
            
        plt.title('%iD, %i V/cm   $D_L = $%.2f'%(d, field_strength, slope_L) + '   $D_{T_y} = $%.2f'%slope_T1 + '   $D_{T_z} = $%.2f'%slope_T2)
            
        plt.plot(t, sigma_squared_L, label='$\sigma_L^2$')
        plt.plot(t, sigma_squared_T1, label='$\sigma_{T_y}^2$')
        plt.plot(t, sigma_squared_T2, label='$\sigma_{T_z}^2$')
        
        plt.xlabel('t [s]')
        plt.ylabel('$\sigma^2$ [cm$^2$]')
        
        plt.legend()
        
        plt.grid()
        plt.show()
        
```

Our transverse and longitudinal diffusion coefficients are calculated by determining the slopes of the lines above, which show the change in positional variation over time for a group of 100 electrons. The values we should check against are available at https://lar.bnl.gov/properties/index.html#e-trans.

For our own reference, plots of distributions in position are displayed below.

```python
for i, d in enumerate([1, 3]):
    for key, value in sorted(file_tree[i].items()):
        group = Graph(value)
        
        for k in range(0, len(group.t[0]), round(len(group.t[0]) / 9)):
            x = []
            y = []
            z = []
            for j in range(group.n):
                x.append(group.x[j][k])
                y.append(group.y[j][k])
                z.append(group.z[j][k])
            
            fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 6))

            plots = [[x, y], [x, z], [y, z]]
            labels = [['x', 'y'], ['x', 'z'], ['y', 'z']]

            for j, subplot in enumerate([ax1, ax2, ax3]):
                subplot.scatter(plots[j][0], plots[j][1], color=colors[i])
                subplot.set_aspect(1.)

                subplot.set_xlabel(labels[j][0] + ' [$\mu$m]')
                subplot.set_ylabel(labels[j][1] + ' [$\mu$m]')

                divider = make_axes_locatable(subplot)
                axHistx = divider.append_axes("top", 1.2, pad=0.1, sharex=subplot)
                axHisty = divider.append_axes("right", 1.2, pad=0.1, sharey=subplot)

                axHistx.xaxis.set_tick_params(labelbottom=False)
                axHisty.yaxis.set_tick_params(labelleft=False)

                binwidth = 1
                xymax = max(np.max(np.abs(plots[j][0])), np.max(np.abs(plots[j][1])))
                lim = (int(20) + 1)*binwidth

                bins = np.arange(-lim, lim + binwidth, binwidth)
                axHistx.hist(plots[j][0], bins=bins, histtype='stepfilled', color=colors[i])
                axHistx.set_title('%iD - %0.3f'%(d, group.t[0][k]) + ' ns')
                axHisty.hist(plots[j][1], bins=bins, orientation='horizontal', histtype='stepfilled', color=colors[i])

                axHistx.set_yticks([0, 5, 10])
                axHisty.set_xticks([0, 5, 10])

            plt.draw()     
            plt.show()
        
```
