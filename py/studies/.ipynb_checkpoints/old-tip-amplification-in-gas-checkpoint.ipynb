```python
import glob
import re

from matplotlib import pyplot as plt
from utils import Graph
import numpy as np
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from collections import OrderedDict

def hit_check(x, y, z):
    h = x
    
    if h >= 0.0:
        return True
    
    return False
```

```python
file_tree = {}

v_per_cm_2 = []
mult_factor_2 = []

top_ionization_list = []

for file in glob.glob('amp-x-check/2/*.txt'):
    key = int(re.search(r'(\d*)(?:V)', file).group(1))
    if (key in file_tree):
        file_tree[key].append(file)
    else:
        file_tree[key] = [file]

for key, value in sorted(file_tree.items()):
    group = Graph(value, smooth=False, cutoff=0.01, order=3)

    ionization_list = [k[-1] for k in group.ionized]
    top_ionization_list.append(ionization_list)
    max_ionized = np.max([k[-1] for k in group.ionized])
    
    est_ionized = np.log(2) * np.array(ionization_list)
    mu = np.mean(est_ionized)
    sigma2 = np.var(est_ionized)
    mult_f = np.exp(mu + 0.5 * sigma2)
    
    multiplication_factor = mult_f
    
    v_per_cm_2.append(key)
    #mult_factor_2.append(multiplication_factor)
    mult_factor_2.append(2**np.mean(ionization_list))
    '''
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 4))
    
    hits = 0

    for i in range(group.n):
        x, y, z = group.x[i], group.y[i], group.z[i]
        if hit_check(x[-1], y[-1], z[-1]) is True:
            hits += 1
            ax1.plot(y, x, color=cm.plasma(group.ionized[i][-1] / max_ionized), alpha=0.6, label=int(group.ionized[i][-1]))
        else:
            ax1.plot(y, x, color='#BBBBBB', alpha=0.6)
        
    # 100 nm: ax1.plot([-1e-4 * 1e6, -1e-7 * 1e6, -0.5e-7 * 1e6, 0, 0.5e-7 * 1e6, 1e-7 * 1e6, 1e-4 * 1e6], [3e-3 * 1e6, 3.5e-3 * 1e6, (3.5e-3 + 0.866025404e-7) * 1e6, (3.5e-3 + 1e-7) * 1e6, (3.5e-3 + 0.866025404e-7) * 1e6, 3.5e-3 * 1e6, 3e-3 * 1e6], color='0.0')
    ax1.plot([-1e-4 * 1e6, -1e-6 * 1e6, -0.5e-6 * 1e6, 0, 0.5e-6 * 1e6, 1e-6 * 1e6, 1e-4 * 1e6], [3e-3 * 1e6, 3.5e-3 * 1e6, (3.5e-3 + 0.866025404e-6) * 1e6, (3.5e-3 + 1e-6) * 1e6, (3.5e-3 + 0.866025404e-6) * 1e6, 3.5e-3 * 1e6, 3e-3 * 1e6], color='0.0')

    ax1.set_xlabel('z Position [$\mu$m]')
    ax1.set_ylabel('x Position [$\mu$m]')
    ax1.set_title('%i V/cm, %0.2f electron multiplication factor'%(key, multiplication_factor))

    ax1.set_xlim(-1e-5 * 1e6, 1e-5 * 1e6)
    ax1.set_ylim(3.49e-3 * 1e6, 3.51e-3 * 1e6)

    handles, labels = ax1.get_legend_handles_labels()
    by_label = OrderedDict(zip([int(j) for j in labels], handles))
    sorted_legend = dict(sorted(by_label.items(), reverse=True))
    ax1.legend(sorted_legend.values(), sorted_legend.keys())
    
    ax2.hist([k[-200:] for k in group.ke], bins=50, rwidth=1.0)
    
    ax2.set_ylim(0, 30)

    ax2.set_xlabel('Energy [eV]')
    ax2.set_ylabel('Instances')
    #ax2.set_title('%0.2f eV mean, %0.2f eV standard deviation'%(group.eV_mean(200), group.eV_std_dev(200)))

    ax3.hist([2**k for k in ionization_list], [0, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024], histtype='stepfilled')
    ax3.plot()
    ax3.set_xlim(1, 1e3)
    ax3.set_xscale('log')
    
    plt.show()
    '''
```

```python
from scipy.stats import lognorm

for temp_list in top_ionization_list:
    test_list = np.log(2) * np.array(temp_list)

    mu = np.mean(test_list)
    sigma = np.std(test_list)

    plt.plot(np.logspace(0, 3, 100), lognorm.pdf(np.logspace(0, 3, 100), sigma, scale=np.exp(mu)))
    plt.hist(np.exp(test_list), bins=[0, 2, 4, 8, 16, 32, 64, 128], density=True, histtype='step')
    plt.ylim(0, 1)
    plt.xlim(1, 100)
    plt.plot([np.exp(mu + 0.5*sigma**2), np.exp(mu + 0.5*sigma**2)], [0, 1])
    plt.xscale('log')
    plt.show()
```
