```python
import glob
import re
from math import sqrt, floor, ceil

import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

from utils import Graph

colors = ['C0', 'C1']
field_strength = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]
```

The below two cells load the necessary files and compute the diffusion coefficients.

```python
file_tree = [{}, {}]

for i, d in enumerate([1, 3]):
    for field in field_strength:
        for file in glob.glob(f'lar-diffusion/{d}d/{field}/*.txt'):
            key = int(re.search(r'(\d*)(?:V)', file).group(1))
            if (key in file_tree[i]):
                file_tree[i][key].append(file)
            else:
                file_tree[i][key] = [file]
```

```python
threshold = 3 # Throw out points more than n standard deviations above the mean

D_L = []
D_Ty = []
D_Tz = []

for key, value in sorted(file_tree[1].items()):
    group = Graph(value)

    t_avg = np.mean([k[-1] for k in group.t])

    x_vals = [k[-1] for k in group.x]
    y_vals = [k[-1] for k in group.y]
    z_vals = [k[-1] for k in group.z]

    this_D_L = np.var([x_vals[k] for k in np.where(np.abs(stats.zscore(x_vals)) < threshold)[0]], ddof=1) / t_avg * 1e1
    D_L.append(this_D_L)
    
    this_D_Ty = np.var([y_vals[k] for k in np.where(np.abs(stats.zscore(y_vals)) < threshold)[0]], ddof=1) / t_avg * 1e1
    D_Ty.append(this_D_Ty)
    
    this_D_Tz = np.var([z_vals[k] for k in np.where(np.abs(stats.zscore(z_vals)) < threshold)[0]], ddof=1) / t_avg * 1e1
    D_Tz.append(this_D_Tz)
```

The cell below plots the diffusion coefficients as a function of bulk field strength. [6] refers to https://lar.bnl.gov/properties/index.html#e-trans.

```python
# plt.rcParams['font.size'] = 14
# plt.rcParams['font.family'] = 'serif'
# fig = plt.figure(figsize=(12, 6))

plt.plot(field_strength, D_L, label='$D_L$')
plt.plot(field_strength, D_Ty, label='$D_{T_y}$')
plt.plot(field_strength, D_Tz, label='$D_{T_z}$')
plt.plot(field_strength, [6.9404, 6.5406, 6.0453, 5.5969, 5.2090, 4.8751, 4.5863, 4.3344, 4.1128, 3.9163], '--', label='$D_L$ from [6]')
plt.plot(field_strength, [17.2051, 19.1958, 18.6916, 17.6541, 16.5709, 15.5700, 14.6781, 13.8914, 13.1982, 12.5855], '--', label='$D_T$ from [6]')

plt.xlabel('Electric Field [Vcm$^{-1}$]')
plt.ylabel('Diffusion Coefficient')
plt.title('Diffusion in LAr')

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# plt.savefig('lar_diffusion.pdf', bbox_inches='tight')
```

For our own purposes, we display scatter plots and histograms of snapshots of our simulated electrons.

```python
field_to_plot = 6000

for i, d in enumerate([1, 3]):
    for key, value in sorted(file_tree[i].items()):
        if key != field_to_plot:
            continue
            
        group = Graph(value)
        
        for k in range(0, len(group.t[0]), round(len(group.t[0]) / 8)):
            x = []
            y = []
            z = []
            for j in range(group.n):
                x.append(group.x[j][k])
                y.append(group.y[j][k])
                z.append(group.z[j][k])
            
            fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 6))

            plots = [[x, y], [x, z], [y, z]]
            labels = [['x', 'y'], ['x', 'z'], ['y', 'z']]

            for j, subplot in enumerate([ax1, ax2, ax3]):
                subplot.scatter(plots[j][0], plots[j][1], color=colors[i])
                subplot.set_aspect(1.)

                subplot.set_xlabel(labels[j][0] + ' [$\mu$m]')
                subplot.set_ylabel(labels[j][1] + ' [$\mu$m]')

                divider = make_axes_locatable(subplot)
                axHistx = divider.append_axes("top", 1.2, pad=0.1, sharex=subplot)
                axHisty = divider.append_axes("right", 1.2, pad=0.1, sharey=subplot)

                axHistx.xaxis.set_tick_params(labelbottom=False)
                axHisty.yaxis.set_tick_params(labelleft=False)

                binwidth = 1
                xymax = max(np.max(np.abs(plots[j][0])), np.max(np.abs(plots[j][1])))
                lim = (int(20) + 1)*binwidth

                bins = np.arange(-lim, lim + binwidth, binwidth)
                axHistx.hist(plots[j][0], bins=bins, histtype='stepfilled', color=colors[i])
                axHistx.set_title('%iD - %0.3f'%(d, group.t[0][k]) + ' ns')
                axHisty.hist(plots[j][1], bins=bins, orientation='horizontal', histtype='stepfilled', color=colors[i])

                axHistx.set_yticks([0, 5, 10])
                axHisty.set_xticks([0, 5, 10])

            plt.draw()     
            plt.show()
        
```
